```python
#In last lecture we have studied SMOT
#What was happening in SMOT >> in SMOT you were able to creat synthetic data points
#What was it actually?
#Whenever you creat a synthetic datapoints let's say you have 2 data points and between these two data points you are creating some synthetic data points >>this is called as interpolation
#It is the process of estimatiing the unknown values from known values
#Generating new data points between two known data points
```

```python
#data interpolation >> It is the process of estimating unknown values from the known values
#There are 3 types of interpolation
#1. linear interpolation
#2. cubic interpolation
#3. polynomial interpolation
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9])

plt.scatter(x, y)
```

```python
# to generate synthetic datapoints
# linear interpolation >> linear in nature
x_new  = np.linspace(1, 5, 10) # creat new x values
y_interp = np.interp(x_new, x, y)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
# cubic interpolation
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])

plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
f = interp1d(x, y, kind = 'cubic')
```

```python
f
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
#polynomial interpolation
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 1, 25])

#interpolate using polynomial interpolation
p = np.polyfit(x, y, 2) # degree2
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x_new, y_interp)
```
