```python
import pandas as pd
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
# feature scaling
# 1. Standardisation >> mu = 0, sigma = 1
# normalization (minmax scaling) >> data comes bewteen 0 and 1
# unit vector >> 1
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.distplot(df['total_bill'])
```

```python
# how to do scaling
```

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
  zscore = (i-mean)/std
  std_data.append(zscore)
```

```python
std_data
```

```python
# feature scaling optional and it does not change the distrbutionn of data
```

```python
sns.distplot(std_data) # distribution remains same
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill']]) # it will calculate mu and sigma
```

```python
scaler.transform(df[['total_bill']]) # dataframe format
```

```python
# for multiple columns
scaler.fit_transform(df[['total_bill', 'tip']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill', 'tip']]), columns = ['total_bill', 'tip'])
```

```python
# new data/test data comes
scaler.transform([[13, 12]]) #reresentation of unseen data
```

```python
# fit actually learns from the trained data and if the new data comes you directly use transform
```

```python
# normalization/ minmax scaling

from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['total_bill', 'tip']])
```

```python
min_max.transform([[13, 12]])
```

```python
# unit vector
from sklearn.preprocessing import normalize
```

```python
uv = normalize(df[['total_bill']])
uv # since one dimension so all the values are 1
```
