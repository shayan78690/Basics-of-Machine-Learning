```python
import pandas as pd
import numpy as np

#Creat a sample dataset with some missing values
data = {
    'A' : [1, 2, 100, 4, 5],
    'B' : [3, np.nan, 7, 8, 9],
    'C' : [np.nan, 12, 13, 14, 15],
    'D' : [16, 17, 18, np.nan, 20]
}

#convert the dictionaries to pandas dataframe
df = pd.DataFrame(data)
print(df)
```

```python
#whenever you have normal distribution data and no outlier, impute with mean
```

```python
df['mean_manipulation'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age', 'mean_manipulation']]
```

```python
sns.histplot(df['mean_manipulation'], kde = True)
```

```python
#mean imputation work well with normally distributed data/no outlier is present
```

```python
df['median_imputation'] = df['age'].fillna(df['age'].median())
```

```python
df[['age', 'mean_manipulation', 'median_imputation']]
```
