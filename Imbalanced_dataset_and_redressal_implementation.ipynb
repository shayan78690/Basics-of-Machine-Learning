```python
#Class imbalance - When one class has very high percentage as compared to another class
#Undersampling(Downsampling)
#Oversampling(Upsampling)
#SMOTE(Synthetic minority oversampling technique)
```

```python
import pandas as pd
import numpy as np

np.random.seed(1) # for reproducibility of result
```

```python
no_samples = 1000
class_0_ratio = 0.9
no_class_0 = int(0.9 * no_samples)
no_class_1 = 100
```

```python
no_class_0
```

```python
no_class_1
```

```python
class_0 = {'feature1' : np.random.normal(0, 1, no_class_0),
           'feature2' : np.random.normal(0, 1, no_class_0),
           'target' : [0] * no_class_0
           }
```

```python
class_0
```

```python
class_0 = pd.DataFrame(class_0)
```

```python
class_0
```

```python
class_1 = {'feature1' : np.random.normal(3, 1, no_class_1),
           'feature2' : np.random.normal(3, 1, no_class_1),
           'target' : [1] * no_class_1
           }
```

```python
class_1
```

```python
class_1 = pd.DataFrame(class_1)
```

```python
class_1
```

```python
# we have generated a synthetic data using numpy random module, class_0 for the first dataframe then we have class1 for the second dataframe
```

```python
# combine both dataframe
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df
```

```python
df.target.value_counts()
```

```python
# this is the classical example of class imbalance
```

```python
# upsampling
df_minority = df[df.target == 1]
df_majority = df[df.target == 0]
```

```python
df_minority
```

```python
df_majority
```

```python
# in upsampling you repeat the minority class data point
```

```python
# replace >> if replace is true, it simply means data point can be repeated
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, replace = True, n_samples = len(df_majority), random_state = 1)
# here replace = True means we want to do replacement and n_sample means the number of times you want to repeat and random_state = 1 means values are reproducable
```

```python
df_minority_upsampled
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled.target.value_counts()
```

```python
# downsampling
df_minority
```

```python
df_majority
```

```python
df_majority_downsampled = resample(df_majority, replace = False, n_samples = len(df_minority), random_state = 1)
```

```python
df_majority_downsampled
```

```python
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled
```

```python
df_downsampled.target.value_counts()
```

```python
# SMOTE >> synthetic minority oversampling technique
```

```python
# another way of creating datapoints
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class=1, weights=[0.90], random_state =1)
```

```python
# here make_classification is a function inside sklearn.datasets
# n_samples=1000 means you want 1000 datapoints
# n_redundant=0 means you don't want your columns to be duplicated
# n_features=2 means you want 2 features
# n_clusters_per_class=1 means you want for each class(class0 and class1) you want two clusters
# weights=[0.90] means majority of the class you want 90%
```

```python
X
```

```python
y
```

```python
df1 = pd.DataFrame(X, columns = ['f1', 'f2'])
df2 = pd.DataFrame(y, columns = ['target'])
final_df = pd.concat([df1, df2], axis = 1)
```

```python
final_df
```

```python
# let's visualize this
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c = final_df['target'])
```
