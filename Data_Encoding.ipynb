```python
# Nominal/OHE >> Binary vector for each of the categories
# example single, married, separated
# single: [1, 0, 0]
# married: [0, 1, 0]
# seperated: [0, 0, 1]
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'status': ['single', 'married', 'seperated', 'single', 'married']})
```

```python
df
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['status']]).toarray() # fit will try to learn which are the unique values and then just simply apply
```

```python
encoded
```

```python
encoder.get_feature_names_out()
```

```python
encoder_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
# if new data comes
encoder.transform([['single']]).toarray()
```

```python
# pd.get_dummies() h/w
# same thing you can achive using this function
```

```python
pd.concat([df, encoder_df], axis = 1)
```

```python
# ordinal encoding

# high school : 1
# graduate : 2
# post graduate : 3
# phd : 4
```

```python
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({'qualification' : ["HS", "GR", "PG", "PHD", "HS", "PHD"]})
```

```python
encoder = OrdinalEncoder(categories=[["HS", "GR", "PG", "PHD"]])
```

```python
encoder.fit_transform(df[['qualification']])
```

```python
encoder.transform([['GR']])
```

```python
# target guided ordianl encoding
# encoding based on some relationship with target variables
# used when a lot of uniqye categories are there
# replace the category with mean and median of respective group
```

```python
df = pd.DataFrame({'time':['lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner'],
             'total_bill': [120, 130, 90, 125, 150, 190, 160, 180, 189]})
```

```python
df
```

```python
mean_price = df.groupby('time')['total_bill'].mean().to_dict()
```

```python
df['time_encoded'] = df['time'].map(mean_price)
```
